```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger une image
img = cv2.imread("../../Image radar/2014-09-11T09-08-52-300.png")

# Convertir l'image en niveaux de gris si elle est en couleur
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Appliquer l'algorithme de détection de contours (par exemple, Canny)
edges = cv2.Canny(img_gray, 100, 200)  # Vous pouvez ajuster les seuils ici

# Trouver les contours dans l'image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Créer une copie de l'image pour dessiner les contours
img_contours = np.copy(img)

# Colorier les contours avec différentes couleurs
for i, contour in enumerate(contours):
    cv2.drawContours(img_contours, [contour], -1, (0, 255, 0), 2)

# Afficher l'image avec les contours colorés
plt.imshow(cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB))
plt.title('Contours colorés')
plt.axis('off')
plt.show()
```
