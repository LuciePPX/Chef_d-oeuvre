## **Dans se fichier SobelContour** <br>

On prend en argument une unique image RADAR : *2014-09-11T09-08-52-300.png*<br>

Le but de ce fichier est de **tester l'algorithme Sobel** sur une unique image. <br> 

On ne prend donc pas en considération : le clustering, le temps, les effets géométriques. <br>

L'algorithme va permettre de tracer les contours des tâches contenue dans l'image. 


### **Importations**

```python
import cv2
import matplotlib.pyplot as plt
```

### **Traitement de l'image** <br>

#### Détection contour des tâches

```python

# Charger une image en niveaux de gris
img = cv2.imread('../Image radar/2014-09-11T09-08-52-300.png', cv2.IMREAD_GRAYSCALE)

# Appliquer l'opérateur Sobel pour détecter les contours
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)  # Sobel horizontal
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)  # Sobel vertical
```

#### Visualiser les contours

```python
# Afficher les résultats
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Image originale')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobelx, cmap='gray')
plt.title('Contours horizontaux (Sobel)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobely, cmap='gray')
plt.title('Contours verticaux (Sobel)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## **Conclusion** <br>

Analyse des contours des tâches selon un balayage vertical puis horizontal
